Let's rename some of the columns in the dataframe.
Pandas dataframes have a "rename" property that accept a dictionary of "old name":"new name" pairs.
The *'inplace=True'* command makes sure our changes are permanently applied to the dataframe, instead of just temporarily.

```python
df.rename(columns={'Economy..GDP.per.Capita.':'GDP',
                 'Health..Life.Expectancy.': 'Health',
                  'Trust..Government.Corruption':'Gov_Trust'}, inplace=True)
```
